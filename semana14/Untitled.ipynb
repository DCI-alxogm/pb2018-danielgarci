```python
def recta(a,m,b=0):
    return m*a+b
```

```python
x=np.linspace(-5,5,20)
plt.plot(x,recta(x,3))
plt.plot(x,recta(x,3,-2))
plt.plot(x,recta(x,-(1/3)))
plt.grid()
```

```python
def recta2(a,m=1,b=0,label='x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    return
```

```python
x=np.linspace(-2,2,20)
recta2(x)
recta2(x,-2,-1,'-x+1')
plt.legend()
plt.grid()
```

```python
def recta2(a,m=1,b=0,label='x'):
    if m<0:
        print('Pendiente negativa, se muestra cero')
        m=0
    z=m*a+b
    plt.plot(a,z,label=label)
    return
```

```python
x=np.linspace(-2,2,20)
recta2(x)
recta2(x,-2,-1,'-x+1')
plt.legend()
plt.grid()
```

```python
def recta(a,m,b=0):
    return m*a+b
```

```python
def recta2(a,m=1,b=0,label='x'):
    z=m*a+b
    plt.plot(a,z,label=label)
    return
```

```python
x=np.linspace(-2,2,20)
recta2(x)
recta2(x,-2,-1,'-x+1')
plt.legend()
plt.grid()
```

##Solución de la escuación del movimiento armónico simple, por el método de Euler

```python
V=np.zeros(10000)
O=np.zeros(10000)
h=0.001
W=2
V[0]=0
O[0]=10
```

```python
def va(O,h,W=1):
    return -(O*h*W**2)
def Oa(V,h):
    return V*h
for i in range(9999):
    
    V[i+1]=V[i]+va(O[i],h)
    O[i+1]=O[i]+Oa(V[i],h)
 
t=np.zeros(10000)
t[0]=0.001
for k in range(9999):
    t[k+1]=t[k]+0.001
```

```python
plt.plot(V,O,label='osc')
plt.grid()
plt.legend()
```

```python
plt.plot(t,V,label='osc')
plt.grid()
plt.legend()
```

```python
plt.plot(t,O,label='periodo')
plt.grid()
plt.legend()
```

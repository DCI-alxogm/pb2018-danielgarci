```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as cons
%matplotlib inline
fname='COBE-FIRAS-CMB monopole spectrum.txt'
dtype1=np.dtype([('frecuency','f8'),('FIRAS monopole spectrum MJy/sr','f8'),('residual monopole spectrum kJy/sr','f8'),('spectrum uncertainty kJy/sr','f8'),('modeled galaxy spectrum kJy/sr','f8')])
a=np.loadtxt(fname,dtype=dtype1, usecols=(0,1,2,3,4))
a
frecuency=a['frecuency']
frecuency
FIRASdata=a['FIRAS monopole spectrum MJy/sr']#*(pow(10,-26))
FIRASdata
residualmonopole=a['residual monopole spectrum kJy/sr']
residualmonopole
uncertainty=(a['spectrum uncertainty kJy/sr']/1000)
uncertainty
model=a['modeled galaxy spectrum kJy/sr']#*(pow(10,-26)))/1000
model
plt.plot(frecuency,FIRASdata,label='FIRAS DATA 2.725 K blackbody',color='green')
plt.legend()
plt.grid()
plt.errorbar(frecuency,FIRASdata,uncertainty,fmt='.',ecolor='g',capthick=1)

```

```python
def I(v,T):
    a=v*100*cons.speed_of_light
    b=(cons.h*a)/(cons.Boltzmann*T)
    c=(cons.h*a**3)
    d=(cons.speed_of_light*np.exp(b))
    e=(c/d)/1e-17
    return e
```

```python
i1=I(frecuency,2.3)
i2=I(frecuency,3)
i3=I(frecuency,4)
i4=I(frecuency,5)
plt.plot(frecuency,FIRASdata,label='experimental data',color='blue')
plt.plot(frecuency,i1,color='red',label='2.3 K')
plt.plot(frecuency,i2,color='violet',label='3 K')
plt.plot(frecuency,i3,color='orange',label='4 K')
plt.plot(frecuency,i4,color='black',label='5 K')
plt.grid()
plt.legend()
```
